```python
grouped_by_job_title = df.groupby('job_title')['salary'].mean()

# Group the data by industry and calculate average salary
grouped_by_industry = df.groupby('employment_type')['salary'].mean()

# Group the data by experience level and calculate average salary
grouped_by_experience = df.groupby('experience_level')['salary'].mean()

print("Grouped by Job Title:\n", grouped_by_job_title)
print("\nGrouped by Industry:\n", grouped_by_industry)
print("\nGrouped by Experience:\n", grouped_by_experience)
```
