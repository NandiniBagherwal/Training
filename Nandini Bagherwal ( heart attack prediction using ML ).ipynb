```python
#plot a few graphs to understand the data better

#checking patients by sex
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='target',data=df,hue='sex',palette='Set2')
```

It is not clarified with sex 0 and 1 refer to, so for the sake of this project I'm going to assume that 0 is for women and 1 for men.

```python
#checking composition of patients within each sex

plt.subplot(1,3,1)

female_patients=df[(df['target']==1)&(df['sex']==0)].shape[0] #get no. of female patients
female_non_patients=df.shape[0]-female_patients #rest are non patients

plt.pie([female_patients,female_non_patients],labels=['Patients','Non patients'],colors=['pink','skyblue'],autopct='%1.1f%%')
plt.title('Female data')

plt.subplot(1,3,2)  #repeating for male patients

male_patients=df[(df['target']==1)&(df['sex']==1)].shape[0]
male_non_patients=df.shape[0]-male_patients

plt.pie([male_patients,male_non_patients],labels=['Patients','Non patients'],colors=['pink','skyblue'],autopct='%1.1f%%')
plt.title('Male data')

plt.subplot(1,3,3)  #repeating for all patients

patients=df[(df['target']==1)].shape[0]
non_patients=df.shape[0]-patients

plt.pie([patients,non_patients],labels=['Patients','Non patients'],colors=['pink','skyblue'],autopct='%1.1f%%')
plt.title('Combined data')

plt.show()
```

```python
sns.heatmap(df.corr()) #correlation between different variables as heatmap
```

```python
#see the age distribution in heart patients

sns.displot(x='age',data=df[df['target']==1])
plt.title('Age distribution in heart patients')

plt.show()
```

```python
#make a logistic regression model

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x = df.iloc[:,:df.shape[1]-1].values #getting independent variables
y = df['target'].values #getting output (dependent) variable

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7) #splitting dataset into training (70%) and testing (30%) data

mod=LogisticRegression(max_iter=1000) #create the model
mod.fit(x_train,y_train)
```

```python
#test the model using testing data

y_predicted = mod.predict(x_test)

print(y_test)
print(y_predicted)  #view actual and predicted values
```

```python
#check accuracy metrics of model

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

#confusion matrix
confusion_matrix(y_test, y_predicted)
```

```python
#accuracy score
accuracy_score(y_test,y_predicted)
```

Therefore, the created logistic regression model has an accuracy of 86%.

In the data it tested, it got 124 true positives, 141 true negatives, 20 false positives and 23 false negatives.
